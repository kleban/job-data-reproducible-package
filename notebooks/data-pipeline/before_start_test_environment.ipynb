# Environment Test

Run each cell to verify that all required packages are installed correctly.
A ✅ message means the package is working. A ❌ message means something is missing or broken.

Run this notebook **before** starting the pipeline to confirm your environment is ready.

**Documentation:** [Setup](../../SETUP.md) · [Notebook guide](README.md)


```python
import sys
print(f'Python version: {sys.version}')
```

```python
try:
    import pandas as pd
    print(f'✅ pandas {pd.__version__}')
except ImportError as e:
    print(f'❌ pandas — {e}')
```

```python
try:
    import numpy as np
    print(f'✅ numpy {np.__version__}')
except ImportError as e:
    print(f'❌ numpy — {e}')
```

```python
try:
    import matplotlib
    import matplotlib.pyplot as plt
    print(f'✅ matplotlib {matplotlib.__version__}')
except ImportError as e:
    print(f'❌ matplotlib — {e}')
```

```python
try:
    import seaborn as sns
    print(f'✅ seaborn {sns.__version__}')
except ImportError as e:
    print(f'❌ seaborn — {e}')
```

```python
try:
    from dotenv import load_dotenv
    import os
    import importlib.metadata
    load_dotenv('.env')
    key = os.getenv('OPENAI_API_KEY', 'NOT SET')
    masked = key[:8] + '...' if key not in ('NOT SET', 'YOUR_OPENAI_API_KEY') else key
    version = importlib.metadata.version('python-dotenv')
    print(f'✅ python-dotenv {version} — OPENAI_API_KEY = {masked}')
except ImportError as e:
    print(f'❌ python-dotenv — {e}')
```

```python
try:
    import fast_langdetect
    from fast_langdetect import detect
    result = detect('Hello world')
    print(f'✅ fast-langdetect — detect("Hello world") → {result}')
except ImportError as e:
    print(f'❌ fast-langdetect — {e}')
```

```python
try:
    import tqdm
    print(f'✅ tqdm {tqdm.__version__}')
except ImportError as e:
    print(f'❌ tqdm — {e}')
```

```python
try:
    import torch
    print(f'✅ torch {torch.__version__} — CUDA available: {torch.cuda.is_available()}')
except ImportError as e:
    print(f'❌ torch — {e}')
```

```python
try:
    import transformers
    print(f'✅ transformers {transformers.__version__}')
except ImportError as e:
    print(f'❌ transformers — {e}')
```

```python
try:
    import sentence_transformers
    print(f'✅ sentence-transformers {sentence_transformers.__version__}')
except ImportError as e:
    print(f'❌ sentence-transformers — {e}')
```

```python
try:
    import openai
    print(f'✅ openai {openai.__version__}')
except ImportError as e:
    print(f'❌ openai — {e}')
```

```python
try:
    import rapidfuzz
    from rapidfuzz import fuzz
    score = fuzz.ratio('data science', 'data scientist')
    print(f'✅ rapidfuzz {rapidfuzz.__version__} — fuzz.ratio test score: {score:.1f}')
except ImportError as e:
    print(f'❌ rapidfuzz — {e}')
```
